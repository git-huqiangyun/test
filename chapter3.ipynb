```python
# 大数据 20222146013 胡强云
# 数据加载
import numpy as np
import pandas as pd
import itertools
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.callbacks import EarlyStopping

plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号

# 导入MNIST数据集
mnist = keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# 查看4个NumPy数组的数据形状
print('train_images:', train_images.shape)
print('test_images:', test_images.shape)
print('train_labels:', train_labels.shape)
print('test_labels:', test_labels.shape)
```

```python
# 大数据 20222146013 胡强云
#  对数字图像进行可视化
fig = plt.figure(figsize=(20, 20))
for i in range(14):
    ax = fig.add_subplot(7, 7, i+1)
    ax.imshow(train_images[i], cmap='gray')
    plt.tight_layout()
    ax.set_title("数字:{}".format(train_labels[i]))
plt.show()
```

```python
# 大数据 20222146013 胡强云
# 数据预处理
# 标准化处理
train_x_normalize = train_images / 255.
test_x_normalize = test_images / 255.

# 利用tensorflow的reshape函数将图像数据从二维的28×28转为一维的784的数据
train_x = tf.reshape(train_x_normalize, [train_x_normalize.shape[0], train_x_normalize.shape[1]*train_x_normalize.shape[2]])
test_x = tf.reshape(test_x_normalize, [test_x_normalize.shape[0], test_x_normalize.shape[1]*test_x_normalize.shape[2]])

# 标签数据预处理
train_y = tf.one_hot(train_labels, depth=10)
test_y = tf.one_hot(test_labels, depth=10)
```

```python
# 大数据 20222146013 胡强云
    # 构建网络
# 构建模型
model = keras.Sequential([
    keras.layers.Dense(units=256, input_shape=(784,), activation='relu'),
    keras.layers.Dense(units=128, activation='relu'),
    keras.layers.Dense(units=10, activation='softmax')
])
model.summary()
```

```python
# 大数据 20222146013 胡强云
    # 编译网络
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
# 大数据 20222146013 胡强云
    # 训练网络
# 设置参数
epochs = 100
validation_split = 0.2

# 定义回调函数
earlystop_callback = EarlyStopping(monitor='val_accuracy', min_delta=0.00001, patience=3)

# 训练模型
history = model.fit(train_x, train_y, epochs=epochs, verbose=2, validation_split=validation_split, callbacks=[earlystop_callback])
```

```python
# 大数据 20222146013 胡强云
    # 训练评估
model.evaluate(test_x,test_y, verbose=2)
```

```python
# 大数据 20222146013 胡强云
    # 训练预测
# 对测试集中的前三个样本进行测试
y_proba = model.predict(test_x[:3])
y_proba.round(3)
```

```python
# 大数据 20222146013 胡强云
# 对测试集所有样本进行预测并查看混淆矩阵
test_y_pred = np.argmax(model.predict(test_x), axis=-1)
# 查看混淆矩阵
confusion_mtx = confusion_matrix(test_labels, test_y_pred)
print(confusion_mtx)
```

```python
# 大数据 20222146013 胡强云
    # 自定义可视化函数
# 自定义可视化函数
def plot_confusion_matrix(cm, classes, normalize=False, title='混淆矩阵', cmap=plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i,j], horizontalalignment="center", color="white" if cm[i,j] > thresh else "black")
    plt.tight_layout()
    plt.ylabel('实际标签')
    plt.xlabel('预测标签')

plt.show()

# 对混淆矩阵进行可视化
plot_confusion_matrix(confusion_mtx, classes=range(10))
```

```python
# 大数据 20222146013 胡强云
# 整个模型保存到当前目录中
def saving_the_model(model, model_name='model_history.h5'):
    model.save(model_name)
    print('Saved trained model at %s' % model_name)

saving_the_model(model, 'model_history.h5')

```

```python
# 大数据 20222146013 胡强云
    # 自定义绘制学习曲线的可视化函数
def draw_model(training):
    plt.figure()
    plt.subplot(2, 1, 1)
    plt.plot(training.history['accuracy'])   # history.history.keys()
    plt.plot(training.history['val_accuracy'])
    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='lower right')

    plt.subplot(2, 1, 2)
    plt.plot(training.history['loss'])
    plt.plot(training.history['val_loss'])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper right')
    
    plt.tight_layout()
    plt.show()

# 绘制每个训练周期的学习曲线
draw_model(history)

# 大数据 20222146013 胡强云
```

```python
# 大数据 20222146013 胡强云
# 读取CSV数据
import pandas as pd
MasterFile=pd.read_csv('./FoodScore.csv') 
print(MasterFile.shape)
MasterFile[0:5]
```

```python
# 大数据 20222146013 胡强云
# 绘制直方图
MasterFile.hist()
```

```python
# 大数据 20222146013 胡强云
# 分离因变量
import numpy as np
FileNames=MasterFile['ID']
N=len(FileNames)
Y=np.array(MasterFile['score']).reshape([N,1])
#Y=(Y-np.mean(Y))/np.std(Y)
```

```python
# 大数据 20222146013 胡强云
# 处理图像数据
from PIL import Image

IMSIZE=128
X=np.zeros([N,IMSIZE,IMSIZE,3])
for i in range(N):
    MyFile=FileNames[i]
    Im=Image.open('./case3-food/data/'+MyFile+'.jpg')
    Im=Im.resize([IMSIZE,IMSIZE])
    Im=np.array(Im)/255
    X[i,]=Im
```

```python
# 大数据 20222146013 胡强云
# 展示前10张图像
from matplotlib import pyplot as plt
plt.figure()
fig,ax=plt.subplots(2,5)
fig.set_figheight(7.5)
fig.set_figwidth(15)
ax=ax.flatten()
for i in range(10):
    ax[i].imshow(X[i,:,:,:])
    ax[i].set_title(str(np.round(Y[i],2))) #书本没有对标题数值转换，会有警告提醒
```

```python
# 大数据 20222146013 胡强云
# 划分训练集和测试集，各占50%固定随机种子
from sklearn.model_selection import train_test_split
X0,X1,Y0,Y1=train_test_split(X,Y,test_size=0.5,random_state=0)
```

```python
# 大数据 20222146013 胡强云
# 构建线性回归模型
from keras.layers import Dense, Flatten, Input
from keras import Model
input_layer=Input([IMSIZE,IMSIZE,3])
x=input_layer
x=Flatten()(x)
x=Dense(1)(x)
output_layer=x
model=Model(input_layer,output_layer)
model.summary()
```

```python
# 大数据 20222146013 胡强云
# 编译模型
from keras.optimizers import Adam
model.compile(loss='mse',optimizer=Adam(learning_rate=0.001),metrics=['mse'])
```

```python
# 大数据 20222146013 胡强云
# 模型拟合和预测
model.fit(X0,Y0,
          validation_data=[X1,Y1],
          batch_size=100,
          epochs=100)

#代码示例：3-20
MyPic=Image.open('mypic.jpg')
MyPic
MyPic=MyPic.resize((IMSIZE,IMSIZE))        
MyPic=np.array(MyPic)/255                  
MyPic=MyPic.reshape((1,IMSIZE,IMSIZE,3))   
model.predict(MyPic)   
```
